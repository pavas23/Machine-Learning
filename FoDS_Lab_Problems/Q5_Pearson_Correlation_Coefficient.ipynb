```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('housing_data.csv')
df.head()
```

```python
mean = df.mean()
std = df.std()
df = (df- mean)/std
df.head()
```

```python
y = df['MEDV'].to_numpy()
x = df.drop(['MEDV'],axis=1)
```

```python
features = df.copy()
```

```python
corr = features.corr()
corr
```

```python
corr_vals = corr["MEDV"]
corr_vals = dict(corr_vals)
print(corr_vals)
```

```python
corr_vals.items()
```

```python
for key in corr_vals:
    corr_vals[key] = abs(corr_vals[key])
```

```python
sorted_dict = sorted(corr_vals.items(),key = lambda x: x[1],reverse=True)
```

```python
sorted_dict
```

```python
# taking top k features
def select_features(k):
    columns = []
    count = 0
    for key in sorted_dict:
        if count == k:
            break
        columns.append(key)
        count += 1

    return columns
```

```python
select_features(3)
```

```python
def LinearRegressionModel(X,y):
    # adding a column of bias to feature dataset
    ones = np.array([1 for i in range(len(X[:,0]))])
    ones = ones.reshape(len(X[:,0]),1)
    X = np.concatenate((ones,X),axis=1)
    
    # finding theta's using normal equation
    X_T = X.T
    X_TX = X_T.dot(X)
    theta = np.linalg.inv(X_TX).dot(X_T).dot(y)
    
    return theta
```
