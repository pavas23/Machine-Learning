### Naive Bayes Classifier


A Naive Bayes classifier is a probabilistic machine learning model thatâ€™s used for classification task. The crux of the classifier is based on the Bayes theorem.

<a href="https://towardsdatascience.com/naive-bayes-classifier-81d512f50a7c"><img src="https://miro.medium.com/v2/resize:fit:1020/format:webp/1*tjcmj9cDQ-rHXAtxCu5bRQ.png" title="source: imgur.com" /></a>


Using Bayes theorem, we can find the probability of A happening, given that B has occurred. Here, B is the evidence and A is the hypothesis. The assumption made here is that the predictors/features are independent. That is presence of one particular feature does not affect the other. Hence, it is called naive.

# Step 1: Importing Libraries

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

# Step 2: Creating a Synthetic Dataset

```python
# Creating the classification dataset with one informative feature and one cluster per class
nb_samples = 300
X, y = make_classification(
    n_samples=nb_samples,
    n_features=2,
    random_state=1,
    n_informative=2,
    n_redundant=0
)
```

# Step 3: Data Visualization and Analysis

```python
# Plotting the dataset
plt.figure(figsize=(7.50, 3.50))
plt.subplots_adjust(bottom=0.05, top=0.9, left=0.05, right=0.95)
plt.subplot(111)
plt.scatter(X[:, 0], X[:, 1], marker="o", c=y, s=40, edgecolor="k")
plt.show()
```

# Step 4: Model Fitting

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3
)
```

## Fitting data on Naive Bayes Classifier

```python
from sklearn.naive_bayes import BernoulliNB

# Build a Benroulli Classifier
classifier = BernoulliNB()

# Model training
classifier.fit(X_train, y_train)
```

# Step 5: Model Evaluation

## Running for a single datapoint as an example

```python
# Predict Output For a single datapoint
predicted = classifier.predict([X_test[6]])

print("Actual Value:", y_test[6])
print("Predicted Value:", predicted[0])
```

## Running classifier on all datapoints in Test Set

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)

y_pred = classifier.predict(X_test)
accuray = accuracy_score(y_pred, y_test)

print("The Accuracy obtained is:", accuray)
```

## Visualizing Predictions and their Accuracies

```python
# The following confusion matrix shall show the numbers of Points in the classes
# and their predicted classes as predicted by the NB classifier

labels = [0,1]
cm = confusion_matrix(y_test, y_pred, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```
